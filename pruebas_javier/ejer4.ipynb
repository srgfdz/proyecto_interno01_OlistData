Calcula y representa el numero de reviews por estado y el score medio en cada una de ellas, como ya hemos tenido en cuenta los pedidos con retraso en la seccion anterior vamos a eliminar estos datos del apartado de score ya que se entiende que la opinion será negativa por la tardanza de llegada del producto.

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Obtención del csv olist_order_customer_dataset
olist_order_reviews_dataset_clean = pd.read_csv ('../datasets_limpios/olist_order_reviews_dataset_clean.csv')
olist_order_reviews_dataset_clean
```

```python
# Obtención del csv olist_orders_dataset
olist_orders_dataset = pd.read_csv ('../datasets_limpios/olist_orders_dataset.csv')
olist_orders_dataset
```

```python
# Obtención del csv olist_order_customer
olist_order_customer = pd.read_csv ('../datasets_limpios/olist_order_customer.csv')
olist_order_customer
```

```python
# Merge de los 3 datasets a utilizar
merged_dataset = olist_order_reviews_dataset_clean.merge(olist_orders_dataset,on='order_id').merge(olist_order_customer,on='customer_id')
merged_dataset

merge_tablas = pd.merge(olist_order_customer, olist_orders_dataset, on="customer_id")

# pasar a formato datetime las fechas
merge_tablas["order_delivered_customer_date"] = pd.to_datetime(merge_tablas["order_delivered_customer_date"])
merge_tablas["order_estimated_delivery_date"] = pd.to_datetime(merge_tablas["order_estimated_delivery_date"])

# calcular retraso
merge_tablas["dias_retraso"] = (merge_tablas["order_delivered_customer_date"] - merge_tablas["order_estimated_delivery_date"]).dt.days

# filtrar los que llegan tarde
pedidos_tarde = merge_tablas[merge_tablas["dias_retraso"] > 0]
merged_dataset
```

```python
merged_dataset_filtrado = merged_dataset[~merged_dataset['customer_id'].isin(pedidos_tarde['customer_id'])]
merged_dataset_filtrado
```

```python
# Ordenar reviews por estado
reviews_estado = merged_dataset_filtrado.groupby("customer_state")["review_score"].count()
reviews_estado
```

```python
# ordenar
reviews_estado = reviews_estado.sort_values(ascending=False)

# Gráfico
reviews_estado.plot(kind="bar", figsize=(12, 6), color='skyblue')

# Etiquetas
plt.xlabel("Estado")
plt.ylabel("Cantidad de reseñas")
plt.title("Cantidad de reseñas por estado")
plt.xticks(rotation=90)

# Mostrar
plt.show()
```

```python
# Score media de reviews
reviews_estado_media_score = merged_dataset_filtrado.groupby("customer_state")["review_score"].mean()
reviews_estado_media_score
```

```python
# Ordenar datos de mayor a menor
reviews_estado_media_score = reviews_estado_media_score.sort_values(ascending=False)

# Gráfico de tarta
reviews_estado_media_score.plot(kind="pie", figsize=(8, 8), autopct='%1.1f%%', colors=plt.cm.Paired.colors)

# Etiquetas
plt.ylabel("")
plt.title("Puntuación media de las reviews por estado")

# Mostrar
plt.show()
```
